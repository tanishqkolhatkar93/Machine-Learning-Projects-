```python
import albumentations as A
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import os
```

```python
def plot_examples(images, bboxes=None):
    fig = plt.figure(figsize=(20, 20))
    columns = 6
    rows = 20
    for i in range(1, len(images)):
        img = images[i-1]
        fig.add_subplot(rows, columns, i)
        plt.imshow(img)
    plt.show()
```

```python
basepath = 'flags/'
collect = []
for entry in os.listdir(basepath):
    file = os.path.join(basepath, entry)
    if os.path.isfile(file):
        image = Image.open(file)
        collect.append(image)  
plot_examples(collect)
```

Applying transformations

```python
transform = A.Compose(
    [
        A.VerticalFlip(p=0.6),
        A.HorizontalFlip(p=0.6),
        A.Rotate(limit=30, p=0.6),
        A.InvertImg(),
        A.Blur(blur_limit=3, p=0.5),
        A.RGBShift(r_shift_limit=25, g_shift_limit=25, b_shift_limit=25, p=0.01),
    ]
)
```

```python
def plot_examples_2(images, bboxes=None):
    fig = plt.figure(figsize=(10, 10))
    columns = 4
    rows = 5
    for i in range(1, len(images)):
        img = images[i-1]
        fig.add_subplot(rows, columns, i)
        plt.imshow(img)
    plt.show()

    
```

```python
for j in collect:
    x = np.array(j)
    image_list=[j]
    for i in range(20):
        augmentations = transform(image=x)
        augmented_img = augmentations['image']
        image_list.append(augmented_img)
    plot_examples_2(image_list)

# for x in collect:
#     h = np.array(x)
#     augmentations = transform(image=h)
#     augmented_img = augmentations['image']
#     image_list.append(augmented_img)
```
